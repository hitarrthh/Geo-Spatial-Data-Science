```python
import matplotlib.pyplot as plt
import geopandas as gpd
```

Loading the India Shapefile

```python
shapefile_path = 'india_st.shp'
india_states_gdf = gpd.read_file(shapefile_path)
india_states_gdf = india_states_gdf.rename(columns={'STATE': 'State_Name'})

print("India states shapefile loaded successfully.")
```

Inspecting the Data

```python
print(f"CRS: {india_states_gdf.crs}\n")

print("Columns:", india_states_gdf.columns)
print("First 5 rows:")
print(india_states_gdf.head())
print("\n")
geometry_types = india_states_gdf.geom_type.unique()
print(f"Geometry Types: {geometry_types}")
```

Ploting the State Boundaries

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

india_states_gdf.plot(ax=ax, color='skyblue', edgecolor='black')

ax.set_title("India State Boundaries", fontsize=16)
ax.set_axis_off()

plt.show()
```

Calculateing and Displaying the Area of Each State

```python
india_states_gdf['area_sqkm'] = india_states_gdf.geometry.area / 10**6

print("Calculated Area of Each State:")
print(india_states_gdf[['State_Name', 'area_sqkm']].sort_values(by='State_Name').to_string())
```
